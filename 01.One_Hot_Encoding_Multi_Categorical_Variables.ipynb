## One-Hot encoding for Multi Categorical Variables

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('DataSets/Mercedes Data.csv',
                 usecols= ("X0","X1","X2","X3","X4","X5"))
df.head()
```

##### Value count for X1 column

```python
df.X1.value_counts().sort_values(ascending = False)
```

##### Let's consider top 10 values

```python
top_10 = df.X1.value_counts().sort_values(ascending = False).head(10).index
top_10
```

##### Extra 10 columns for most frequent variable
label 1 for top 10 values and label 0 for other values

```python
for label in top_10:
     df[label] = np.where(df['X1']==label,1,0)
```

```python
df
```

##### Lets Create Functioin for One Hot Encoding

```python
def Ohe_MV(feature,n):
    top_n = df[feature].value_counts().sort_values(ascending = False).head(n).index
    for label in top_n:
        df[label] = np.where(df[feature]==label,1,0)
    return df    
```
